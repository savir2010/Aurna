```python
import pandas as pd
import re

# Load the CSV fill

#AF7 (Muse) → F7 (your dataset).

# AF8 (Muse) → F8 (your dataset).

# TP9 (Muse) → T3 (your dataset).

# TP10 (Muse) → T4 (your dataset).
electrodes = ['T3', 'T4', 'F7', 'F8']  # Replace with your desired 4 channels

# Define frequency bands
bands = ['delta', 'theta', 'alpha', 'beta', 'highbeta', 'gamma']

# Initialize lists to store relevant column names
power_columns = []
coherence_columns = []

# Generate regex patterns for power columns
for band in bands:
    for electrode in electrodes:
        # Format: AB.{band}.{letter}.{electrode}
        pattern = re.compile(fr'^AB\..*\.{band}\.\w+\.{electrode}$')
        matched_columns = [col for col in df.columns if pattern.match(col)]
        if matched_columns:
            print(f"Power columns for {band} and {electrode}: {matched_columns}")
            power_columns.extend(matched_columns)

# Generate regex patterns for coherence columns
for band in bands:
    for i, electrode1 in enumerate(electrodes):
        for j, electrode2 in enumerate(electrodes):
            if i < j:  # Ensure unique pairs
                # Format: COH.{letter}.{band}.{letter}.{electrode1}.{letter}.{electrode2}
                pattern = re.compile(fr'^COH\.[A-Za-z]+\.{band}\.[A-Za-z]+\.{electrode1}\.[A-Za-z]+\.{electrode2}$')
                matched_columns = [col for col in df.columns if pattern.match(col)]
                if matched_columns:
                    print(f"Coherence columns for {band}, {electrode1}, and {electrode2}: {matched_columns}")
                    coherence_columns.extend(matched_columns)

# Combine power and coherence columns
relevant_columns = power_columns + coherence_columns

# Check if all relevant columns exist in the DataFrame
if not relevant_columns:
    raise KeyError("No matching columns found. Check the column names in your CSV file.")

# Filter the dataset to keep only relevant columns
cleaned_df = df[relevant_columns]
clean_df = cleaned_df
# Save the cleaned dataset to a new CSV file
output_file = 'cleaned_data_4channels.csv'
cleaned_df.to_csv(output_file, index=False)
print(f"Cleaned dataset saved to {output_file}")
```

```python

X = df.drop(columns=['disorder'])
y = df['disorder']

label_encoder = LabelEncoder()
y = label_encoder.fit_transform(y)

scaler = StandardScaler()
X = scaler.fit_transform(X)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.5, random_state=42)

# Reshape data to have 61 timesteps and 1 feature
X_train = X_train.reshape((X_train.shape[0], X_train.shape[1], 1))  # (samples, timesteps, 1)
X_test = X_test.reshape((X_test.shape[0], X_test.shape[1], 1))

# Step 2: Build the Model
```

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, Conv1D, MaxPooling1D, LSTM, Dense, Flatten, concatenate, BatchNormalization, Dropout, Attention, GlobalAveragePooling1D
from tensorflow.keras.regularizers import l2
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau

# Step 2: Build the Model
input_shape = (X_train.shape[1], 1)  # (timesteps, 1)

# Input layer
input_layer = Input(shape=input_shape)

# CNN branch
conv1 = Conv1D(filters=512, kernel_size=3, activation='relu', padding='same')(input_layer)
batch_norm1 = BatchNormalization()(conv1)
pool1 = MaxPooling1D(pool_size=2)(batch_norm1)
dropout1 = Dropout(0.5)(pool1)

conv2 = Conv1D(filters=256, kernel_size=3, activation='relu', padding='same')(dropout1)
batch_norm2 = BatchNormalization()(conv2)
pool2 = MaxPooling1D(pool_size=2)(batch_norm2)
dropout2 = Dropout(0.5)(pool2)

conv3 = Conv1D(filters=128, kernel_size=3, activation='relu', padding='same')(dropout2)
batch_norm3 = BatchNormalization()(conv3)
pool3 = MaxPooling1D(pool_size=2)(batch_norm3)
dropout3 = Dropout(0.5)(pool3)

flatten = Flatten()(dropout3)

# RNN branch with Attention
lstm1 = LSTM(512, return_sequences=True, kernel_regularizer=l2(0.01))(input_layer)
batch_norm4 = BatchNormalization()(lstm1)
dropout4 = Dropout(0.5)(batch_norm4)

lstm2 = LSTM(256, return_sequences=True, kernel_regularizer=l2(0.01))(dropout4)
batch_norm5 = BatchNormalization()(lstm2)
dropout5 = Dropout(0.5)(batch_norm5)

lstm3 = LSTM(128, return_sequences=True, kernel_regularizer=l2(0.01))(dropout5)
batch_norm6 = BatchNormalization()(lstm3)
dropout6 = Dropout(0.5)(batch_norm6)

# Attention layer
attention = Attention()([dropout6, dropout6])

# Apply Global Average Pooling to reduce sequence dimension
avg_pool = GlobalAveragePooling1D()(attention)

# Concatenate CNN and RNN (with Attention) outputs
merged = concatenate([flatten, avg_pool])

# Fully connected layers
dense1 = Dense(512, activation='relu', kernel_regularizer=l2(0.01))(merged)
batch_norm7 = BatchNormalization()(dense1)
dropout7 = Dropout(0.5)(batch_norm7)

dense2 = Dense(256, activation='relu', kernel_regularizer=l2(0.01))(dropout7)
batch_norm8 = BatchNormalization()(dense2)
dropout8 = Dropout(0.5)(batch_norm8)

dense3 = Dense(128, activation='relu', kernel_regularizer=l2(0.01))(dropout8)
batch_norm9 = BatchNormalization()(dense3)
dropout9 = Dropout(0.5)(batch_norm9)

# Output layer
output_layer = Dense(1, activation='sigmoid')(dropout9)

# Create the model
model = Model(inputs=input_layer, outputs=output_layer)

# Compile the model with a more adaptive learning rate
model.compile(optimizer='Nadam', loss='binary_crossentropy', metrics=['accuracy'])

# Callbacks
early_stopping = EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True)
reduce_lr = ReduceLROnPlateau(monitor='val_loss', factor=0.5, patience=5, min_lr=1e-6)

# Print model summary
model.summary()
```

```python
# Step 3: Train the Model
early_stopping = EarlyStopping(
    monitor='val_accuracy',  # Monitor validation accuracy
    patience=20,             # Wait 20 epochs before stopping
    restore_best_weights=True  # Restore weights from the best epoch
)

# ReduceLROnPlateau callback
reduce_lr = ReduceLROnPlateau(
    monitor='val_accuracy',  # Monitor validation accuracy
    factor=0.1,              # Reduce learning rate by 0.1
    patience=10,             # Wait 10 epochs before reducing LR
    min_lr=1e-7              # Minimum learning rate
)

history = model.fit(
    X_train, y_train,
    epochs=50,  # Train for a large number of epochs
    batch_size=32,
    validation_split=0.2
)

# Step 4: Evaluate the Model
loss, accuracy = model.evaluate(X_test, y_test)
print(f"Test Accuracy: {accuracy * 100:.2f}%")
```
