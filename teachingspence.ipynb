```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
import numpy as np
```

```python
df = pd.read_csv("Sleep_health_and_lifestyle_dataset.csv")
```

```python
df
```

```python
df.columns
```

```python
df = df[["Sleep Duration", "Stress Level"]]
```

```python
df
```

```python
X = df.drop("Stress Level", axis=1)
```

```python
y = df["Stress Level"]
```

```python
X
```

```python
y
```

```python
type(y)
```

```python
type(X)
```

```python
model = LinearRegression()
model.fit(X, y)
```

```python
y_pred = model.predict(X)
```

```python
y_pred
```

```python
plt.scatter(X, y, color='blue', label="Actual")
plt.plot(X, y_pred, color='red', label="Prediction")
plt.legend()
plt.show()
```

```python
m = model.coef_[0]  # Slope
b = model.intercept_  # Intercept
f"Formula: y = {m:.2f}x+ {b:.2f}"
```

```python
sleep_duration = np.array([[7.5]])
predicted_stress = model.predict(sleep_duration)
print(f"Predicted stress for Sleep Duration = 7.5: {predicted_stress[0]:.2f}")
```
